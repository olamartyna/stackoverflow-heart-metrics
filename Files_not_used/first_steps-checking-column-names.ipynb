```python
import xml.etree.ElementTree as ET
```

```python
file_path = "datasets/Users.xml"

# Look at the first 5 rows to inspect the structure
for event, elem in ET.iterparse(file_path, events=("start",)):
    if elem.tag == "row":
        print(elem.attrib)  # Prints all available columns
        break  # Exit after one row to check structure
```

```python
def check_column_names(table):
    for event, elem in ET.iterparse(table, events=("start",)):
        if elem.tag == "row":
            print(elem.attrib)  # Prints all available columns
            break  # Exit after one row to check structure
```

```python
users = "datasets/Users.xml"
check_column_names(users)
```

```python
posts = "datasets/Posts.xml"
check_column_names(posts)
```

```python
tags = "datasets/Tags.xml"
check_column_names(tags)
```

```python
votes = "datasets/Votes.xml"
check_column_names(votes)
```

```python
comments = "datasets/Comments.xml"
check_column_names(comments)
```
