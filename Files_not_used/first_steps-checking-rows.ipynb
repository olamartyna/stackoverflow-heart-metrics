### First I will check if data is present in datasets.

```python
import pandas as pd
import xml.etree.ElementTree as ET
```

I am going to check how many rows Tags have and what are column names

```python
def count_rows(file_path):
    count = 0
    for event, elem in ET.iterparse(file_path, events=("start",)):
        if elem.tag == "row":
            count += 1
    return count
```

```python
datasets = ["Tags.xml"]
for dataset in datasets:
    path = f"datasets/{dataset}"
    print(f"{dataset}: {count_rows(path)} rows")
```
