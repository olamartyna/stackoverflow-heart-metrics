```python
import sqlite3
```

```python
conn = sqlite3.connect("stackoverflow.db")
```

```python
tables = conn.execute("SELECT name FROM sqlite_master WHERE type='table';").fetchall()
for (table,) in tables:
    count = conn.execute(f"SELECT COUNT(*) FROM {table};").fetchone()[0]
    print(f"{table}: {count} rows")
```
