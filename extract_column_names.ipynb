```python
import xml.etree.ElementTree as ET
```

```python
def get_column_names(xml_file):
    for event, elem in ET.iterparse(xml_file, events=("start",)):
        if elem.tag == "row":
            column_names = list(elem.attrib.keys())  # Get keys in order
            elem.clear()  # Free memory
            return column_names  # Return ordered column names
    print("Column Names in Order:", column_names)
```

```python
users = "datasets/Users.xml"
get_column_names(users)
```

```python
comments = "datasets/Comments.xml"
get_column_names(comments)
```

```python
posts = "datasets/Posts.xml"
get_column_names(posts)
```

```python
tags = "datasets/Tags.xml"
get_column_names(tags)
```

```python
votes = "datasets/Votes.xml"
get_column_names(votes)
```
