5. Null / missing values

```python
def null_counts(table):
    df = pd.read_sql(f"SELECT * FROM {table} LIMIT 10000", conn)
    return df.isnull().sum()
```

```python
null_counts("Users")
```

```python
null_counts("Comments")
```

```python
null_counts("Posts")
```

```python
null_counts("Tags")
```

```python
null_counts("Votes")
```
